### 5.2. Preparing dataset

Due to the complex and violatile stock market and various trading restrictions, stock prices observed are noisy. Simultaneously, financial time series is non-stationary and exhibits the overlapping of useful signals and noise, which makes traditional denoising methods ineffective.

Therefore, a wavelet transform with multi-scale characteristics to denoise the dataset by using it to separate the useful signal from the noise. Specifically, the coif3 wavelet function with 3 decomposition layers.
